
## Gegevensanalyse en visualisatie

Vandaag gaan we ook een kijkje nemen naar gegevensanalyse en visualisatie met behulp van de pandas en matplotlib bibliotheken.

We beginnen met het inlezen van gegevens uit csv-bestanden en het berekenen van het totale bedrag dat elke leverancier heeft uitgegeven:

```python
import pandas as pd

# Lees de facturen.csv in een DataFrame
df_facturen = pd.read_csv('C:/Users/Martijnbosch/cumlaude.ai/cumlaude.ai - Documents/Academy/Python/PyCursusFacturen.csv')

# Bereken het totale bedrag per leverancier
total_amount_per_leverancier = df_facturen.groupby('LeverancierID')['Bedrag'].sum()
```

Vervolgens willen we de namen van de leveranciers weergeven in plaats van de leverancier-ID's. Deze informatie komt uit een ander bestand, leveranciers.csv:

```python
df_leveranciers = pd.read_csv('C:/Users/Martijnbosch/cumlaude.ai/cumlaude.ai - Documents/Academy/Python/PyCursusLeveranciers.csv')
leverancier_names = df_leveranciers.set_index('LeverancierID')['LeverancierNaam']
```

Nu kunnen we een staafdiagram maken om de totale bedragen per leverancier te visualiseren:

```python
import matplotlib.pyplot as plt

# Maak het staafdiagram
plt.bar(leverancier_names, total_amount_per_leverancier)
plt.xlabel('Leverancier')
plt.ylabel('Totaal Bedrag')
plt.title('Totaal Bedrag per Leverancier')
plt.xticks(rotation=45, ha='right')  # Draai de labels op de x-as voor betere leesbaarheid
plt.tight_layout()

# Toon het staafdiagram
plt.show()
```
